# Creating Linear Regression Model

```python
crop_data = pd.get_dummies(data=crop)
crop_data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = crop_data.drop('Production',axis=1)
X.head()
```

```python
y = crop_data['Production']
y.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
```

```python
crop_model = LinearRegression()
```

```python
crop_model.fit(X_train,y_train)
```

```python
crop_predictions = crop_model.predict(X_test)
crop_predictions
```

```python
crop_model.coef_
```

```python
crop_model.intercept_
```

```python
predicted_crop_val = pd.DataFrame({'Actual':y_test,'Predicted':crop_predictions})
predicted_crop_val
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
crop['Production'].mean()
```

```python
crop_predictions.mean()
```

```python
mean_absolute_error(y_test,crop_predictions)
```

```python
mean_squared_error(y_test,crop_predictions)
```

```python
np.sqrt(mean_squared_error(y_test,crop_predictions))
```

```python
r = r2_score(y_test,crop_predictions)
print("R2score to predict using Linear Regression is ",r)
```

# Decision Tree Regressor model

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(min_samples_split=10, max_depth=20)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score
print('MSE:', mean_squared_error(y_pred, y_test))
print("MAE :", mean_absolute_error(y_pred, y_test))
print('R2 score', r2_score(y_pred, y_test))
```

```python
y_pred = model.predict(X_train)
print('MSE:', mean_squared_error(y_pred, y_train))
print("MAE :", mean_absolute_error(y_pred, y_train))
print('R2 score', r2_score(y_pred, y_train))
```

ML Models

```python
Draft = pd.get_dummies(data=Draft)
Draft
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = Draft.drop('Draft (KN)',axis=1)
X.head()
```

```python
X.shape
```

```python
y = Draft['Draft (KN)']
y.head()
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
sd= StandardScaler()
X_train = sd.fit_transform(X_train)
X_test =sd.transform(X_test)
```

model1 = Decision Tree

```python
from sklearn.tree import DecisionTreeRegressor
model1 = DecisionTreeRegressor(min_samples_split=4, max_depth=16, max_features = 13, random_state=42)
```

```python
model1.fit(X_train, y_train)
```

```python
y_pred = model1.predict(X_test)
y_pred_tr = model1.predict(X_train)
print('MSE on Train :', mean_squared_error(y_train, y_pred_tr))
print('RMSE on Train :', mean_squared_error(y_train, y_pred_tr, squared=False))
print('R2 on train :', model1.score(X_train, y_train)) 
print('MSE on test:', mean_squared_error(y_test, y_pred))
print('RMSE on test:', mean_squared_error(y_test, y_pred, squared=False))
print('R2 score on test:', r2_score(y_test, y_pred))
```

Model2 = Random forest

```python
from sklearn.ensemble import RandomForestRegressor
model2 = RandomForestRegressor()
```

```python
model2.fit(X_train, y_train)
```

```python
from sklearn.model_selection import GridSearchCV
param = {'n_estimators':[60, 80, 90,100,110.120],
        'min_samples_split':[4,6,8,10,12,14],
        'max_depth': [5,6,7,8,9,10,12]}
```

```python
grid_s = GridSearchCV(estimator=model2, param_grid = param, cv = 5, scoring='neg_mean_squared_error')
grid_s.fit(X_train, y_train)
```

```python
grid_s.best_params_
```

```python
y_pred = model2.predict(X_test)
y_pred_tr = model2.predict(X_train)
print('MSE on Train :', mean_squared_error(y_train, y_pred_tr))
print('RMSE on Train :', mean_squared_error(y_train, y_pred_tr, squared=False))
print('R2 on train :', model1.score(X_train, y_train)) 
print('MSE on test:', mean_squared_error(y_test, y_pred))
print('RMSE on test:', mean_squared_error(y_test, y_pred, squared=False))
print('R2 score on test:', r2_score(y_test, y_pred))
```
